```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd

df = pd.read_csv('data/AB_NYC_2019.csv')
df = df[df['price']<=500]
df.head(5)
```

```python
from shapely.geometry import Point
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
gdf.head(5)
```

```python
from IPython.display import display, Markdown

display(Markdown('# AirBnB Listings Price Map'))

#Plot map of NYC
nyc = gpd.read_file(gpd.datasets.get_path('nybb'),encoding='utf-8')
nyc = nyc.to_crs(epsg=4326)
ax = nyc.plot(figsize=(20, 20), alpha=0.5, edgecolor='k')
#Plot points from airbnb listings dataframe
gdf.plot(ax=ax, column='price', cmap='PiYG_r', markersize = 5.0)
plt.show()
```
